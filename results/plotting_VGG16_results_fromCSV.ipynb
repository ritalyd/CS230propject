```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#with open("../VGG16_results.csv", 'r') as csvfile:

df = pd.read_csv('VGG16_results.csv', delimiter = ';')
```

```python
df.head()
```

```python
train_acc = df['acc']
train_f1 = df['f1']
train_loss = df['loss']
val_acc = df['val_acc']
val_f1 = df['val_f1']
val_loss = df['val_loss']
```

```python
# plot train and validation Accuracy vs 100 Epochs
plt.figure(1)
line1 = plt.plot(range(1,101),train_acc,'r',label='train_acc')
line2 = plt.plot(range(1,101), val_acc, 'b', label='val_acc')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
# plt.legend(handles=[line1,line2])
plt.title('VGG-16 train validation accuracy vs 100 epoch')
plt.savefig('plot_VGG16_train_valid_accuracy_100epoch')
plt.show()
```

```python
# plot train and validation f1 score vs 100 Epochs
plt.figure(2)
line1 = plt.plot(range(1,101),train_f1,'r',label='train_f1')
line2 = plt.plot(range(1,101), val_f1, 'b', label='val_f1')
plt.ylabel('F1 Score')
plt.xlabel('Epoch')
# plt.legend(handles=[line1,line2])
plt.title('VGG-16 train validation f1 score vs 100 epoch using')
plt.savefig('plot_VGG16_train_valid_f1_100epoch')
plt.show()
```

```python
# plot train and validation Loss vs 100 Epochs
plt.figure(3)
line1 = plt.plot(range(1,101),train_loss,'r',label='train_loss')
line2 = plt.plot(range(1,101), val_loss, 'b', label='val_loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
# plt.legend(handles=[line1,line2])
plt.title('VGG-16 train validation loss vs 100 epoch using Adam lr0001')
plt.savefig('plot_VGG16_train_valid_loss_100epoch')
plt.show()
```
